```python
#Make copy of accounts_all do aggregations on variables and then merge on train and test
import pandas as pd
customer_panel_train = pd.read_csv("/home/jovyan/shared/datasets/RetailBanking/customer_panel_train.csv")
customer_panel_test = pd.read_csv("/home/jovyan/shared/datasets/RetailBanking/customer_panel_test.csv")
customers_all = pd.read_csv("/home/jovyan/shared/datasets/RetailBanking/customers_all.csv")
accounts_agg_all = pd.read_csv("agg_acc_all.csv")
    
merged_train = (
        customer_panel_train
        .merge(accounts_agg_all, left_on="CustomerID", right_on="CustomerID", how="left")
        .merge(customers_all, left_on = "CustomerID", right_on = "CustomerID", how = "left")
        )

merged_test = (
        customer_panel_test
        .merge(accounts_agg_all, left_on="CustomerID", right_on="CustomerID", how="left")
        .merge(customers_all, left_on = "CustomerID", right_on = "CustomerID", how = "left")
        )

merged_train['HomeCity'] = merged_train.groupby('HomeCity')['DefaultLabel'].transform('mean') 
city_mean_map = merged_train.groupby('HomeCity')['DefaultLabel'].mean()
merged_test['HomeCity'] = merged_train['HomeCity'].map(city_mean_map)
    
merged_train.to_csv(f'customer_panel_train_merge.csv')
merged_test.to_csv(f'customer_panel_test_merge.csv')
```
